```python
#Change the data type of Date and time column to align with the dtype also to reduce the memory usage of the int64 dtypes
#checking for counts of null/NA is the dataset
sp_df['Date'] = pd.to_datetime(sp_df['Date'])
sp_df['Time'] = pd.to_datetime(sp_df['Time'])
sp_df['RowID'] = sp_df['RowID'].astype('int16')
sp_df['Quantity'] = sp_df['Quantity'].astype('int16')
sp_df['Year in City'] = sp_df['Year in City'].astype('int16')
```

```python
#The updated Sales Table will be 
Salesdf = sp_df[['RowID', 'InvoiceID', 'Branch', 'City','CustomerType', 'Gender', 'UnitPrice','Quantity', 'Tax5Percent', 'Date', 'Time', 'Payment', 'CostOfGoodsSold', 'GrossMarginPercentage', 'GrossIncome', 'Rating', 'YearInCity', 'Age', 'ProdID']]
```

```python
Salesdf.head()
```

```python
#Create Connection to the SQL Server
conn = py.connect(Driver= 'SQL Server',
                      Server='ORINDU\SQLEXPRESS',
                      Database='Sales_Analysis',
                      Trusted_Connection = 'yes')
```

```python
cursor = conn.cursor()
```

```python
cursor.execute ('''
                CREATE TABLE Salestbl(
                   RowID int, 
                   InvoiceID nvarchar(50) primary key, 
                   Branch nvarchar(50), 
                   City nvarchar(50),
                   CustomerType nvarchar(50),
                   Gender nvarchar(50),
                   UnitPrice float,
                   Quantity int, 
                   Tax5Percent float, 
                   Date datetime, 
                   Time datetime, 
                   Payment nvarchar(50), 
                   CostOfGoodsSold float, 
                   GrossMarginPercentage float, 
                   GrossIncome float, 
                   Rating float, 
                   YearInCity int,
                   Age nvarchar(50),
                   ProdID nvarchar(50)
                  )
            ''')
```

```python
#Isert the Sales Dataframe into the sales_table
for row in Salesdf.itertuples():
    cursor.execute('''
        INSERT INTO Salestbl(RowID, InvoiceID, Branch, City, CustomerType, Gender, UnitPrice, Quantity, Tax5Percent, Date, Time, Payment, CostOfGoodsSold, GrossMarginPercentage, GrossIncome, Rating, YearInCity, Age, ProdID)
        VALUES(?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)
        ''',
                row.RowID, 
                row.InvoiceID, 
                row.Branch, 
                row.City,
                row.CustomerType,
                row.Gender,
                row.UnitPrice,
                row.Quantity, 
                row.Tax5Percent, 
                row.Date, 
                row.Time, 
                row.Payment, 
                row.CostOfGoodsSold, 
                row.GrossMarginPercentage, 
                row.GrossIncome, 
                row.Rating, 
                row.YearInCity,
                row.Age,
                row.ProdID
                )
cursor.commit()
```
